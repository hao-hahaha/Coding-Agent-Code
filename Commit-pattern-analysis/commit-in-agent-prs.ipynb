```python
# ========= imports =========
import os
import shutil
import subprocess
import re
import json
from typing import List, Dict

# ========= helpers =========
def api_to_clone_url(api_url: str) -> str:
    return api_url.replace("https://api.github.com/repos/", "https://github.com/")

def parse_pr_number_from_html(html_url: str) -> int:
    m = re.search(r"/pull/(\d+)", html_url)
    if not m:
        raise ValueError(f"Cannot parse PR number from {html_url}")
    return int(m.group(1))


# ========= core: get commit timeline for ONE PR =========
def get_pr_commit_timeline_via_pull_ref(
    merge_sha: str,
    pr_number: int,
    tmp_dir: str,
    fetch_depth: int = 500,
):
    subprocess.run(
        ["git", "fetch", "--depth", str(fetch_depth), "origin", merge_sha],
        cwd=tmp_dir,
        check=True,
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL,
    )
    subprocess.run(
        ["git", "checkout", "--detach", "FETCH_HEAD"],
        cwd=tmp_dir,
        check=True,
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL,
    )

    # base at merge time
    base_sha = subprocess.check_output(
        ["git", "rev-parse", f"{merge_sha}^1"],
        cwd=tmp_dir,
        text=True
    ).strip()

    # fetch PR head
    pr_ref = f"pull/{pr_number}/head"
    subprocess.run(
        ["git", "fetch", "--depth", str(fetch_depth), "origin", pr_ref],
        cwd=tmp_dir,
        check=True,
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL,
    )

    pr_head_sha = subprocess.check_output(
        ["git", "rev-parse", "FETCH_HEAD"],
        cwd=tmp_dir,
        text=True
    ).strip()

    # list commits (oldest -> newest)
    log_output = subprocess.check_output(
        [
            "git", "log",
            "--reverse",
            "--pretty=format:%H%x09%an%x09%ae%x09%s",
            f"{base_sha}..{pr_head_sha}",
        ],
        cwd=tmp_dir,
        text=True,
    )

    timeline = []
    for idx, line in enumerate(log_output.splitlines(), start=1):
        sha, author_name, author_email, message = line.split("\t", 3)
        timeline.append({
            "commit_index": idx,
            "sha": sha,
            "author_name": author_name,
            "author_email": author_email,
            "message": message,
        })

    return timeline


# ========= repo-level runner =========
def get_commit_timelines_for_repo(
    api_repo_url: str,
    pr_list: List[dict],
    tmp_dir: str = "_tmp_pr_commit_timelines",
):
    clone_url = api_to_clone_url(api_repo_url)

    if os.path.exists(tmp_dir):
        shutil.rmtree(tmp_dir)
    os.makedirs(tmp_dir, exist_ok=True)

    try:
        subprocess.run(
            ["git", "init"],
            cwd=tmp_dir,
            check=True,
            stdout=subprocess.DEVNULL,
        )
        subprocess.run(
            ["git", "remote", "add", "origin", clone_url],
            cwd=tmp_dir,
            check=True,
            stdout=subprocess.DEVNULL,
        )

        timelines = {}

        for pr in pr_list:
            merge_sha = pr.get("merge_commit_sha")
            html_url = pr.get("html_url")
            pr_order = pr.get("pr_order_in_repo")

            if not merge_sha or not html_url or not pr_order:
                continue

            pr_number = parse_pr_number_from_html(html_url)

            try:
                timeline = get_pr_commit_timeline_via_pull_ref(
                    merge_sha=merge_sha,
                    pr_number=pr_number,
                    tmp_dir=tmp_dir,
                )
                timelines[pr_order] = {
                    "pr_number": pr_number,
                    "html_url": html_url,
                    "merge_sha": merge_sha,
                    "timeline": timeline,
                }
            except Exception as e:
                print(f"PR_{pr_order} (#{pr_number}) failed: {e}")
                timelines[pr_order] = None

        return timelines

    finally:
        shutil.rmtree(tmp_dir, ignore_errors=True)
        
def repo_already_done(repo_out_dir: str) -> bool:
    return os.path.exists(os.path.join(repo_out_dir, "_DONE"))

# # ========= RUN HERE =========
# api_repo = "https://api.github.com/repos/514-labs/moose"
# repo_slug = "514-labs_moose"
# pr_list = repo_pr_records_clean[api_repo]

# OUTPUT_ROOT = "commit-timelines"
# repo_out_dir = os.path.join(OUTPUT_ROOT, repo_slug)
# os.makedirs(repo_out_dir, exist_ok=True)

# timelines = get_commit_timelines_for_repo(api_repo, pr_list)

# # ========= PRINT + SAVE =========
# for pr_order in sorted(timelines):
#     data = timelines[pr_order]
#     if not data:
#         continue

#     timeline = data["timeline"]
#     pr_number = data["pr_number"]
#     html_url = data["html_url"]
#     merge_sha = data["merge_sha"]

#     print(f"\n================ PR_{pr_order} ({len(timeline)} commits) ================")
#     print(f"PR link: {html_url}")

#     for c in timeline:
#         print(
#             f"[{c['commit_index']}] "
#             f"{c['sha'][:8]}  "
#             f"{c['author_name']} <{c['author_email']}>"
#         )
#         print(f"    {c['message']}")

#     # ---- save JSON ----
#     out_path = os.path.join(repo_out_dir, f"PR_{pr_order}.json")
#     with open(out_path, "w", encoding="utf-8") as f:
#         json.dump(
#             {
#                 "repo": repo_slug.replace("_", "/"),
#                 "pr_order": pr_order,
#                 "pr_number": pr_number,
#                 "html_url": html_url,
#                 "merge_sha": merge_sha,
#                 "commits": timeline,
#             },
#             f,
#             indent=2,
#             ensure_ascii=False,
#         )

#     print(f"    💾 Saved to {out_path}")
```

```python
# ========= BATCH RUN: all repos =========
OUTPUT_ROOT = "commit-timelines"
os.makedirs(OUTPUT_ROOT, exist_ok=True)

for api_repo, pr_list in repo_pr_records_clean.items():
    repo_slug = api_repo.replace(
        "https://api.github.com/repos/", ""
    ).replace("/", "_")

    repo_out_dir = os.path.join(OUTPUT_ROOT, repo_slug)

    # ⏭️ skip if DONE
    if repo_already_done(repo_out_dir):
        print(f"⏭️  Skip repo (DONE): {repo_slug}")
        continue

    print(f"\n📦 Processing repo: {repo_slug}")
    os.makedirs(repo_out_dir, exist_ok=True)

    try:
        timelines = get_commit_timelines_for_repo(api_repo, pr_list)

        for pr_order in sorted(timelines):
            data = timelines[pr_order]
            if not data:
                continue

            timeline = data["timeline"]
            pr_number = data["pr_number"]
            html_url = data["html_url"]
            merge_sha = data["merge_sha"]

            print(f"  🔹 PR_{pr_order}: {len(timeline)} commits")

            out_path = os.path.join(repo_out_dir, f"PR_{pr_order}.json")
            with open(out_path, "w", encoding="utf-8") as f:
                json.dump(
                    {
                        "repo": api_repo.replace(
                            "https://api.github.com/repos/", ""
                        ),
                        "pr_order": pr_order,
                        "pr_number": pr_number,
                        "html_url": html_url,
                        "merge_sha": merge_sha,
                        "commits": timeline,
                    },
                    f,
                    indent=2,
                    ensure_ascii=False,
                )

        # ✅ mark DONE
        done_path = os.path.join(repo_out_dir, "_DONE")
        with open(done_path, "w") as f:
            f.write("OK\n")

        print(f"  ✅ Finished repo: {repo_slug} (DONE)")

    except Exception as e:
        print(f"  ❌ Failed repo {repo_slug}: {e}")
        continue
```
