```python
from collections import Counter
import os

ext_counter = Counter()

for files in pr_full_data_clean["files"]:
    if not isinstance(files, list):
        continue

    for f in files:
        if isinstance(f, dict):
            filename = f.get("filename")
        else:
            continue

        if not filename:
            continue

        _, ext = os.path.splitext(filename)
        ext = ext.lower() if ext else "<no_ext>"
        ext_counter[ext] += 1

ext_counter.most_common(10)

# top 10 file extension in merged agent pull requests
```

```python
# find python and typescript pull requests

TARGET_EXTS = {".py", ".ts", ".tsx"}

def pr_hit_exts(files, target_exts):
    hits = set()
    if not isinstance(files, list):
        return hits

    for f in files:
        if isinstance(f, dict):
            filename = f.get("filename")
        else:
            continue

        if not filename:
            continue

        _, ext = os.path.splitext(filename)
        if ext.lower() in target_exts:
            hits.add(ext.lower())

    return hits

pr_full_data_clean["hit_exts"] = pr_full_data_clean["files"].apply(
    lambda files: pr_hit_exts(files, TARGET_EXTS)
)
```

Python PRs Security Analysis

```python
python_pr_df = pr_full_data_clean[
    pr_full_data_clean["hit_exts"].apply(lambda exts: ".py" in exts)
].copy()

python_pr_df
```

```python
pr_df_repo = pr_df[["html_url", "repo_url"]].drop_duplicates()

python_pr_df = python_pr_df.merge(
    pr_df_repo,
    on="html_url",
    how="left"
)
```
