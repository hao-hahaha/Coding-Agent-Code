```python
import requests
from urllib.parse import urlparse
import base64

GITHUB_TOKEN = ""

def get_merge_commit_sha(pr_url, github_token=GITHUB_TOKEN):
    """
    Given a GitHub PR URL, return merge_commit_sha and merged_at
    """
    # Parse URL: https://github.com/{owner}/{repo}/pull/{number}
    parts = urlparse(pr_url).path.strip("/").split("/")
    owner, repo, _, pr_number = parts

    api_url = f"https://api.github.com/repos/{owner}/{repo}/pulls/{pr_number}"

    headers = {}
    if github_token:
        headers["Authorization"] = f"Bearer {github_token}"

    resp = requests.get(api_url, headers=headers)
    resp.raise_for_status()

    data = resp.json()

    return {
        "owner": owner,
        "repo": repo,
        "pr_number": pr_number,
        "merged": data["merged"],
        "merged_at": data["merged_at"],
        "merge_commit_sha": data["merge_commit_sha"],
    }

def get_files_changed_by_commit(owner, repo, sha, github_token=GITHUB_TOKEN):
    url = f"https://api.github.com/repos/{owner}/{repo}/commits/{sha}"
    headers = {}
    if github_token:
        headers["Authorization"] = f"Bearer {github_token}"

    resp = requests.get(url, headers=headers)
    resp.raise_for_status()
    data = resp.json()

    files = []
    for f in data.get("files", []):
        files.append({
            "filename": f["filename"],
            "status": f["status"],        # added / modified / removed
            "additions": f["additions"],
            "deletions": f["deletions"],
            "changes": f["changes"],
        })
    return files

def get_file_at_commit(owner, repo, filepath, sha, github_token=GITHUB_TOKEN):
    url = f"https://api.github.com/repos/{owner}/{repo}/contents/{filepath}"
    params = {"ref": sha}

    headers = {}
    if github_token:
        headers["Authorization"] = f"Bearer {github_token}"

    resp = requests.get(url, headers=headers, params=params)
    resp.raise_for_status()

    data = resp.json()

    # GitHub contents API returns base64
    if data.get("encoding") == "base64":
        return base64.b64decode(data["content"]).decode("utf-8", errors="ignore")
    else:
        raise RuntimeError("Unexpected content encoding")
```

```python
def safe_get_merge_commit_sha(pr_url, github_token=GITHUB_TOKEN):
    try:
        info = get_merge_commit_sha(pr_url, github_token)
        return info
    except requests.HTTPError as e:
        return {
            "error": str(e),
            "merge_commit_sha": None,
            "merged": False,
        }
    except Exception as e:
        return {
            "error": str(e),
            "merge_commit_sha": None,
            "merged": False,
        }
```

```python
# 1. use html url get pull request merge commit sha
# 2. use merge commit sha get changed file
# 3. use changed files get source code in file level

import numpy as np
import pandas as pd

df = pd.read_parquet("merged_pr_df.parquet")

N_SHARDS = 8
df["shard_id"] = np.arange(len(df)) % N_SHARDS

for i in range(N_SHARDS):
    shard_df = df[df["shard_id"] == i].copy()
    shard_df.to_parquet(f"merged_pr_df_shard_{i}.parquet", index=False)
    print(f"Shard {i}: {len(shard_df)} PRs")
```
