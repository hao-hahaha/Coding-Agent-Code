```python
def extract_file_names(files):
    if not isinstance(files, list):
        return []
    names = []
    for f in files:
        if isinstance(f, dict):
            name = f.get("filename")
            if name:
                names.append(name)
    return names

# get files names and then get languages
pr_full_data["files_name"] = pr_full_data["files"].apply(extract_file_names)
```

```python
def filename_to_language(filename, lang_map):
    """
    filename: e.g. 'src/main.test.ts'
    return: language string or None
    """
    import os
    ext = os.path.splitext(filename)[1].lower()  # '.ts'
    return lang_map.get(ext)

def files_name_to_languages(files_name, lang_map):
    """
    files_name: list[str]
    return: list[str] (deduplicated languages)
    """
    if not isinstance(files_name, list):
        return []

    languages = set()
    for fname in files_name:
        if not isinstance(fname, str):
            continue
        lang = filename_to_language(fname, lang_map)
        if lang:
            languages.add(lang)

    return sorted(languages)
```

```python
LANG_MAP = {
    # Python
    ".py": "python",

    # JavaScript
    ".js": "javascript",
    ".jsx": "javascript",

    # TypeScript
    ".ts": "typescript",
    ".tsx": "typescript",

    # Go
    ".go": "go",

    # Java
    ".java": "java",

    # C#
    ".cs": "csharp",

    # C / C++
    ".c": "c",
    ".cpp": "cpp",

    # Rust
    ".rs": "rust",

    # PHP
    ".php": "php",
}
```

```python
pr_full_data["languages"] = pr_full_data["files_name"].apply(
    lambda files: files_name_to_languages(files, LANG_MAP)
)

pr_full_data["is_multilang"] = pr_full_data["languages"].apply(lambda x: len(x) > 1)

pr_full_data
```

```python
import json
import re
import hashlib
import subprocess
from pathlib import Path
from typing import Optional, Tuple

import pandas as pd


# ============================================================
# 0) Environment checks
# ============================================================
def _check_cmd(cmd: str) -> str:
    p = subprocess.run([cmd, "--version"], capture_output=True, text=True)
    if p.returncode != 0:
        raise RuntimeError(f"Command not available or failed: {cmd}\nSTDERR:\n{p.stderr}")
    return (p.stdout.strip() or p.stderr.strip()).strip()

print("Node:", _check_cmd("node"))
print("npm:", _check_cmd("npm"))
print("npx:", _check_cmd("npx"))


# ============================================================
# 1) Helpers
# ============================================================
def safe_slug(s: str, max_len: int = 80) -> str:
    raw = str(s)
    slug = re.sub(r"[^a-zA-Z0-9._-]+", "_", raw).strip("_")
    h = hashlib.md5(raw.encode("utf-8")).hexdigest()[:8]
    slug = slug[:max_len]
    return f"{slug}_{h}"

def get_context(code: str, line: int, radius: int = 3) -> Optional[str]:
    """Return a small snippet around 1-based line number."""
    if not isinstance(code, str) or not code:
        return None
    lines = code.splitlines()
    if not line or line < 1 or line > len(lines):
        return None
    start = max(1, line - radius)
    end = min(len(lines), line + radius)
    snippet = []
    for ln in range(start, end + 1):
        prefix = ">> " if ln == line else "   "
        snippet.append(f"{prefix}{ln:4d}: {lines[ln-1]}")
    return "\n".join(snippet)

def ensure_workspace(eslint_root: Path) -> None:
    eslint_root.mkdir(parents=True, exist_ok=True)

def ensure_eslint_config(eslint_root: Path) -> Path:
    """
    Create ESLint v9 flat config in CommonJS form:
      eslint.config.cjs
    This avoids ESM / "type": "module" issues.
    """
    config_path = eslint_root / "eslint.config.cjs"
    if config_path.exists():
        return config_path

    config_text = r"""
/**
 * ESLint v9 Flat Config (CommonJS)
 * - JS/JSX only
 * - Enables plugin rule sets when available
 */
const sonarjs = require("eslint-plugin-sonarjs");
const unicorn = require("eslint-plugin-unicorn");
const importPlugin = require("eslint-plugin-import");

function rulesFrom(plugin, configName) {
  try {
    const cfg = plugin && plugin.configs && plugin.configs[configName];
    return (cfg && cfg.rules) ? cfg.rules : {};
  } catch (e) {
    return {};
  }
}

module.exports = [
  {
    files: ["**/*.{js,jsx}"],
    languageOptions: {
      ecmaVersion: "latest",
      sourceType: "module",
      parserOptions: {
        ecmaFeatures: { jsx: true },
      },
    },
    plugins: {
      sonarjs,
      unicorn,
      import: importPlugin,
    },
    rules: {
      // Plugin "recommended" rules (if present)
      ...rulesFrom(sonarjs, "recommended"),
      ...rulesFrom(unicorn, "recommended"),
      ...rulesFrom(importPlugin, "recommended"),

      // A few pragmatic defaults (optional; comment out if you want pure recommended-only)
      "no-unused-vars": "warn",
      "no-undef": "error",
    },
  },
];
""".lstrip()

    config_path.write_text(config_text, encoding="utf-8")
    return config_path

def ensure_npm_deps(eslint_root: Path, auto_install: bool = False) -> None:
    """
    If auto_install=True, will:
      npm init -y
      npm install -D eslint eslint-plugin-sonarjs eslint-plugin-unicorn eslint-plugin-import
    """
    if not auto_install:
        return

    pkg = eslint_root / "package.json"
    if not pkg.exists():
        subprocess.run(["npm", "init", "-y"], cwd=eslint_root, check=True)

    # Install dev deps
    subprocess.run(
        [
            "npm", "install", "-D",
            "eslint",
            "eslint-plugin-sonarjs",
            "eslint-plugin-unicorn",
            "eslint-plugin-import",
        ],
        cwd=eslint_root,
        check=True
    )

def run_eslint(js_path: Path, eslint_root: Path) -> Tuple[bool, list, str, int]:
    """
    Run ESLint on a single JS / JSX file.
    Notes:
      - ESLint returns non-zero if issues exist; we still parse JSON.
      - --no-error-on-unmatched-pattern avoids "No files matching" crashes.
    """
    proc = subprocess.run(
        [
            "npx", "eslint",
            str(js_path),
            "--format", "json",
            "--no-error-on-unmatched-pattern",
        ],
        cwd=eslint_root,
        capture_output=True,
        text=True
    )

    reports = []
    ok = False
    try:
        if proc.stdout.strip():
            reports = json.loads(proc.stdout)
            ok = True
    except Exception:
        ok = False

    return ok, reports, proc.stderr, proc.returncode


# ============================================================
# 2) Parse ESLint outputs
# ============================================================
def parse_eslint_file_row(eslint_reports, html_url, filename, tmp_path):
    """
    One row per file (summary).
    """
    row = {
        "html_url": html_url,
        "filename": filename,
        "tmp_path": tmp_path,

        "eslint_error": 0,
        "eslint_warning": 0,
        "eslint_fatal": 0,
        "eslint_fixable_error": 0,
        "eslint_fixable_warning": 0,

        # plugin hit counters (optional)
        "eslint_sonarjs": 0,
        "eslint_unicorn": 0,
        "eslint_import": 0,
        "eslint_core": 0,
    }

    for fr in eslint_reports or []:
        row["eslint_error"] += fr.get("errorCount", 0)
        row["eslint_warning"] += fr.get("warningCount", 0)
        row["eslint_fatal"] += fr.get("fatalErrorCount", 0)
        row["eslint_fixable_error"] += fr.get("fixableErrorCount", 0)
        row["eslint_fixable_warning"] += fr.get("fixableWarningCount", 0)

        for msg in fr.get("messages", []):
            rule_id = msg.get("ruleId")
            if not rule_id:
                continue
            plugin = rule_id.split("/", 1)[0] if "/" in rule_id else "core"
            key = f"eslint_{plugin}"
            if key in row:
                row[key] += 1
            else:
                # keep a minimal fallback bucket
                row["eslint_core"] += 1

    row["eslint_issues_total"] = row["eslint_error"] + row["eslint_warning"]
    return row

def parse_eslint_issue_rows(eslint_reports, html_url, filename, code, language="javascript", context_radius=3):
    """
    One row per ESLint message (issue-level), including message + code context.
    """
    rows = []
    for fr in eslint_reports or []:
        file_path = fr.get("filePath")

        for msg in fr.get("messages", []):
            rule_id = msg.get("ruleId") or "unknown"
            plugin = rule_id.split("/", 1)[0] if "/" in rule_id else "core"
            line = msg.get("line")
            end_line = msg.get("endLine") or line

            fix = msg.get("fix") or {}
            rows.append({
                "html_url": html_url,
                "filename": filename,
                "file_path": file_path,
                "language": language,

                "rule_id": rule_id,
                "plugin": plugin,

                "severity": msg.get("severity"),  # 1=warn, 2=error
                "message": msg.get("message"),
                "messageId": msg.get("messageId"),
                "node_type": msg.get("nodeType"),

                "line": line,
                "column": msg.get("column"),
                "end_line": end_line,
                "end_column": msg.get("endColumn"),

                "has_fix": bool(fix),
                "fix_range": fix.get("range"),
                "fix_text": fix.get("text"),

                "context": get_context(code, line, radius=context_radius),
            })
    return rows


# ============================================================
# 3) Analyze all JS/JSX from js_pr_df
# ============================================================
def analyze_all_javascript_files(
    js_pr_df: pd.DataFrame,
    eslint_root: str = "lint_workspace",
    tmp_root: str = "tmp_js_analysis_all",
    limit_prs: int | None = None,
    context_radius: int = 3,
    auto_install_deps: bool = False,
):
    eslint_root = Path(eslint_root).resolve()
    ensure_workspace(eslint_root)
    ensure_npm_deps(eslint_root, auto_install=auto_install_deps)
    cfg = ensure_eslint_config(eslint_root)
    print("ESLint config:", cfg)

    tmp_root = (eslint_root / tmp_root).resolve()
    tmp_root.mkdir(parents=True, exist_ok=True)

    file_rows = []
    issue_rows = []

    df_iter = js_pr_df if limit_prs is None else js_pr_df.head(limit_prs)

    for idx, pr_row in df_iter.iterrows():
        html_url = pr_row.get("html_url")
        files = pr_row.get("files")

        if not isinstance(files, list):
            continue

        pr_dir = tmp_root / safe_slug(html_url or f"row_{idx}")
        pr_dir.mkdir(parents=True, exist_ok=True)

        for f in files:
            if not isinstance(f, dict):
                continue

            filename = f.get("filename")
            code = f.get("source_code")

            if not isinstance(filename, str):
                continue
            lower = filename.lower()
            if not lower.endswith((".js", ".jsx")):
                continue
            if not isinstance(code, str) or not code.strip():
                continue

            out_name = safe_slug(filename)
            ext = ".jsx" if lower.endswith(".jsx") else ".js"
            js_path = pr_dir / f"{out_name}{ext}"
            js_path.write_text(code, encoding="utf-8")

            e_ok, e_reports, e_stderr, e_rc = run_eslint(js_path, eslint_root)

            # file-level
            file_row = parse_eslint_file_row(
                e_reports,
                html_url=html_url,
                filename=filename,
                tmp_path=str(js_path),
            )
            file_row.update({
                "eslint_ok": e_ok,
                "eslint_returncode": e_rc,
                "eslint_stderr": (e_stderr or ""),
            })
            file_rows.append(file_row)

            # issue-level (with context)
            issue_rows.extend(
                parse_eslint_issue_rows(
                    e_reports,
                    html_url=html_url,
                    filename=filename,
                    code=code,
                    language="javascript",
                    context_radius=context_radius
                )
            )

    file_df = pd.DataFrame(file_rows)
    issue_df = pd.DataFrame(issue_rows)

    # PR-level aggregation (from file-level)
    if len(file_df) > 0:
        pr_df = (
            file_df
            .groupby("html_url", as_index=False)
            .agg(
                js_files_count=("filename", "nunique"),
                eslint_issues_total=("eslint_issues_total", "sum"),
                eslint_error=("eslint_error", "sum"),
                eslint_warning=("eslint_warning", "sum"),
                eslint_fatal=("eslint_fatal", "sum"),
                eslint_fixable_error=("eslint_fixable_error", "sum"),
                eslint_fixable_warning=("eslint_fixable_warning", "sum"),
                eslint_sonarjs=("eslint_sonarjs", "sum"),
                eslint_unicorn=("eslint_unicorn", "sum"),
                eslint_import=("eslint_import", "sum"),
                eslint_core=("eslint_core", "sum"),
            )
        )
    else:
        pr_df = pd.DataFrame(columns=["html_url"])

    return file_df, issue_df, pr_df


# ============================================================
# 4) Run
# ============================================================
# 你已经有 js_pr_df 了
print("JavaScript PR count:", len(js_pr_df))

js_file_results, js_issue_results, js_pr_level_results = analyze_all_javascript_files(
    js_pr_df=js_pr_df,
    eslint_root="lint_workspace",
    tmp_root="tmp_js_analysis_all",
    limit_prs=None,            # 先想快速验证可改成 50
    context_radius=3,
    auto_install_deps=False,   # 如果你还没在 lint_workspace npm install，可改 True
)

print("JS file-level rows:", len(js_file_results))
print("JS issue-level rows:", len(js_issue_results))
print("JS PR-level rows:", len(js_pr_level_results))

# Save outputs
js_file_results.to_parquet("javascript_static_analysis_file_level.parquet", index=False)
js_file_results.to_csv("javascript_static_analysis_file_level.csv", index=False)

js_issue_results.to_parquet("javascript_static_analysis_issue_level.parquet", index=False)
js_issue_results.to_csv("javascript_static_analysis_issue_level.csv", index=False)

js_pr_level_results.to_parquet("javascript_static_analysis_pr_level.parquet", index=False)
js_pr_level_results.to_csv("javascript_static_analysis_pr_level.csv", index=False)

print("Saved:",
      "javascript_static_analysis_file_level.*",
      "javascript_static_analysis_issue_level.*",
      "javascript_static_analysis_pr_level.*")

# about 1 hour
```

```python
import json
import pandas as pd

def build_pr_json_objects_include_clean_files(
    js_file_results: pd.DataFrame,
    js_issue_results: pd.DataFrame,
):
    base_files = (
        js_file_results[["html_url", "filename", "tmp_path"]]
        .drop_duplicates(subset=["html_url", "filename"])
        .copy()
    )

    issue_grouped = {}
    if js_issue_results is not None and len(js_issue_results) > 0:
        issue_cols = [
            "html_url", "filename",
            "rule_id", "plugin", "severity",
            "message", "messageId", "node_type",
            "line", "column", "end_line", "end_column",
            "has_fix", "fix_range", "fix_text",
            "context",
        ]
        issue_df = js_issue_results.copy()
        for c in issue_cols:
            if c not in issue_df.columns:
                issue_df[c] = None
        issue_df = issue_df[issue_cols]

        for (html_url, filename), g in issue_df.groupby(["html_url", "filename"], dropna=False):
            issues = []
            for _, r in g.iterrows():
                issues.append({
                    "tool": "eslint",
                    "rule_id": r["rule_id"],
                    "plugin": r["plugin"],
                    "severity": r["severity"],
                    "message": r["message"],
                    "messageId": r["messageId"],
                    "node_type": r["node_type"],
                    "line": r["line"],
                    "column": r["column"],
                    "end_line": r["end_line"],
                    "end_column": r["end_column"],
                    "has_fix": r["has_fix"],
                    "fix_range": r["fix_range"],
                    "fix_text": r["fix_text"],
                    "context": r["context"],
                })
            issue_grouped[(html_url, filename)] = issues

    pr_objects = []
    for html_url, pr_files in base_files.groupby("html_url", dropna=False):
        files_list = []
        for _, fr in pr_files.sort_values("filename").iterrows():
            files_list.append({
                "filename": fr["filename"],
                "tmp_path": fr["tmp_path"],
                "issues": issue_grouped.get((html_url, fr["filename"]), []),
            })
        pr_objects.append({
            "html_url": html_url,
            "files": files_list
        })

    return pr_objects


pr_objects = build_pr_json_objects_include_clean_files(
    js_file_results=js_file_results,
    js_issue_results=js_issue_results,
)

out_path = "javascript_eslint_pr_grouped_results.json"
with open(out_path, "w", encoding="utf-8") as f:
    json.dump(pr_objects, f, ensure_ascii=False, indent=2)

print("Saved:", out_path)
print("PR count:", len(pr_objects))
```
