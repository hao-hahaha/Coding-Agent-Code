```python
def extract_file_names(files):
    if not isinstance(files, list):
        return []
    names = []
    for f in files:
        if isinstance(f, dict):
            name = f.get("filename")
            if name:
                names.append(name)
    return names

# get files names and then get languages
pr_full_data["files_name"] = pr_full_data["files"].apply(extract_file_names)
```

```python
def filename_to_language(filename, lang_map):
    """
    filename: e.g. 'src/main.test.ts'
    return: language string or None
    """
    import os
    ext = os.path.splitext(filename)[1].lower()  # '.ts'
    return lang_map.get(ext)

def files_name_to_languages(files_name, lang_map):
    """
    files_name: list[str]
    return: list[str] (deduplicated languages)
    """
    if not isinstance(files_name, list):
        return []

    languages = set()
    for fname in files_name:
        if not isinstance(fname, str):
            continue
        lang = filename_to_language(fname, lang_map)
        if lang:
            languages.add(lang)

    return sorted(languages)
```

```python
LANG_MAP = {
    # Python
    ".py": "python",

    # JavaScript
    ".js": "javascript",

    # TypeScript
    ".ts": "typescript",
    ".tsx": "typescript",

    # Go
    ".go": "go",

    # Java
    ".java": "java",

    # C#
    ".cs": "csharp",

    # C / C++
    ".c": "c",
    ".cpp": "cpp",

    # Rust
    ".rs": "rust",

    # PHP
    ".php": "php",
}
```

```python
pr_full_data["languages"] = pr_full_data["files_name"].apply(
    lambda files: files_name_to_languages(files, LANG_MAP)
)

pr_full_data["is_multilang"] = pr_full_data["languages"].apply(lambda x: len(x) > 1)

pr_full_data
```

```python
import re
import json
import hashlib
import subprocess
from pathlib import Path

import pandas as pd

def _check_cmd(cmd):
    p = subprocess.run([cmd, "--version"], capture_output=True, text=True)
    if p.returncode != 0:
        raise RuntimeError(f"Command not available or failed: {cmd}\nSTDERR:\n{p.stderr}")
    return p.stdout.strip() or p.stderr.strip()

print("Bandit:", _check_cmd("bandit"))
print("Pylint:", _check_cmd("pylint"))


python_pr_df = python_pr_df
print("Python PR count:", len(python_pr_df))


# helpers
def safe_slug(s: str, max_len: int = 80) -> str:
    s = re.sub(r"[^a-zA-Z0-9._-]+", "_", str(s)).strip("_")
    h = hashlib.md5(str(s).encode("utf-8")).hexdigest()[:8]
    s = s[:max_len]
    return f"{s}_{h}"

def run_bandit(py_path: Path):
    proc = subprocess.run(
        ["bandit", "-f", "json", str(py_path)],
        capture_output=True,
        text=True
    )
    issues = []
    ok = False
    try:
        if proc.stdout.strip():
            report = json.loads(proc.stdout)
            issues = report.get("results", []) or []
            ok = True
    except Exception:
        ok = False
    return ok, issues, proc.stderr, proc.returncode

def run_pylint(py_path: Path):
    proc = subprocess.run(
        ["pylint", str(py_path), "--output-format=json", "--score=no"],
        capture_output=True,
        text=True
    )
    issues = []
    ok = False
    try:
        if proc.stdout.strip():
            issues = json.loads(proc.stdout)
            ok = True
        else:
            ok = False
    except Exception:
        ok = False
    return ok, issues, proc.stderr, proc.returncode

def bandit_counts_by_severity(bandit_issues):
    counts = {"LOW": 0, "MEDIUM": 0, "HIGH": 0}
    for it in bandit_issues or []:
        sev = (it.get("issue_severity") or "").upper()
        if sev in counts:
            counts[sev] += 1
    return counts

def pylint_counts_by_type(pylint_issues):
    counts = {"convention": 0, "refactor": 0, "warning": 0, "error": 0, "fatal": 0}
    for it in pylint_issues or []:
        t = it.get("type")
        if t in counts:
            counts[t] += 1
    return counts

# analyze ALL python files using source_code
def analyze_all_python_files(python_pr_df: pd.DataFrame, tmp_root: str = "tmp_python_analysis_all", limit_prs: int | None = None):
    tmp_root = Path(tmp_root)
    tmp_root.mkdir(parents=True, exist_ok=True)

    rows = []
    df_iter = python_pr_df if limit_prs is None else python_pr_df.head(limit_prs)

    for idx, pr_row in df_iter.iterrows():
        html_url = pr_row.get("html_url")
        files = pr_row.get("files")

        if not isinstance(files, list):
            continue

        pr_dir = tmp_root / safe_slug(html_url or f"row_{idx}")
        pr_dir.mkdir(parents=True, exist_ok=True)

        for f in files:
            if not isinstance(f, dict):
                continue

            filename = f.get("filename")
            code = f.get("source_code")

            # only analyze .py with available full source_code
            if not isinstance(filename, str) or not filename.endswith(".py"):
                continue
            if not isinstance(code, str) or not code.strip():
                continue

            # write to temp file (safe name to avoid path traversal and collisions)
            out_name = safe_slug(filename)
            py_path = pr_dir / f"{out_name}.py"
            py_path.write_text(code, encoding="utf-8")

            # bandit
            b_ok, b_issues, b_stderr, b_rc = run_bandit(py_path)
            b_counts = bandit_counts_by_severity(b_issues)

            # pylint
            p_ok, p_issues, p_stderr, p_rc = run_pylint(py_path)
            p_counts = pylint_counts_by_type(p_issues)

            rows.append({
                "html_url": html_url,
                "filename": filename,
                "tmp_path": str(py_path),

                "bandit_ok": b_ok,
                "bandit_returncode": b_rc,
                "bandit_issues_total": len(b_issues),
                "bandit_high": b_counts["HIGH"],
                "bandit_medium": b_counts["MEDIUM"],
                "bandit_low": b_counts["LOW"],
                "bandit_stderr": (b_stderr or "")[:],

                "pylint_ok": p_ok,
                "pylint_returncode": p_rc,
                "pylint_issues_total": len(p_issues),
                "pylint_error": p_counts["error"],
                "pylint_warning": p_counts["warning"],
                "pylint_refactor": p_counts["refactor"],
                "pylint_convention": p_counts["convention"],
                "pylint_fatal": p_counts["fatal"],
                "pylint_stderr": (p_stderr or "")[:],
            })

    return pd.DataFrame(rows)

python_file_results = analyze_all_python_files(python_pr_df, tmp_root="tmp_python_analysis_all", limit_prs=None)
print("File-level rows:", len(python_file_results))
python_file_results.head()


if len(python_file_results) > 0:
    pr_level_results = (
        python_file_results
        .groupby("html_url", as_index=False)
        .agg(
            python_files_count=("filename", "nunique"),

            bandit_any_fail=("bandit_ok", lambda x: (~x).any()),
            pylint_any_fail=("pylint_ok", lambda x: (~x).any()),

            bandit_issues_total=("bandit_issues_total", "sum"),
            bandit_high=("bandit_high", "sum"),
            bandit_medium=("bandit_medium", "sum"),
            bandit_low=("bandit_low", "sum"),

            pylint_issues_total=("pylint_issues_total", "sum"),
            pylint_error=("pylint_error", "sum"),
            pylint_warning=("pylint_warning", "sum"),
            pylint_refactor=("pylint_refactor", "sum"),
            pylint_convention=("pylint_convention", "sum"),
            pylint_fatal=("pylint_fatal", "sum"),
        )
    )
else:
    pr_level_results = pd.DataFrame(columns=["html_url"])

print("PR-level rows:", len(pr_level_results))
pr_level_results.head()

python_file_results.to_parquet("python_static_analysis_file_level.parquet", index=False)
python_file_results.to_csv("python_static_analysis_file_level.csv", index=False)

pr_level_results.to_parquet("python_static_analysis_pr_level.parquet", index=False)
pr_level_results.to_csv("python_static_analysis_pr_level.csv", index=False)

print("Saved:",
      "python_static_analysis_file_level.parquet/csv,",
      "python_static_analysis_pr_level.parquet/csv")

# about 3.5 hours running time
```
