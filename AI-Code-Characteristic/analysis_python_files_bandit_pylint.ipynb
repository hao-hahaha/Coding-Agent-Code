```python
def filename_to_language(filename, lang_map):
    """
    filename: e.g. 'src/main.test.ts'
    return: language string or None
    """
    import os
    ext = os.path.splitext(filename)[1].lower()  # '.ts'
    return lang_map.get(ext)

def files_name_to_languages(files_name, lang_map):
    """
    files_name: list[str]
    return: list[str] (deduplicated languages)
    """
    if not isinstance(files_name, list):
        return []

    languages = set()
    for fname in files_name:
        if not isinstance(fname, str):
            continue
        lang = filename_to_language(fname, lang_map)
        if lang:
            languages.add(lang)

    return sorted(languages)
```

```python
LANG_MAP = {
    # Python
    ".py": "python",

    # JavaScript
    ".js": "javascript",
    ".jsx": "javascript",

    # TypeScript
    ".ts": "typescript",
    ".tsx": "typescript",

    # Go
    ".go": "go",

    # Java
    ".java": "java",

    # C#
    ".cs": "csharp",

    # C / C++
    ".c": "c",
    ".cpp": "cpp",

    # Rust
    ".rs": "rust",

    # PHP
    ".php": "php",
}
```

```python
import json
import re
import hashlib
import subprocess
from pathlib import Path
from typing import Optional

import pandas as pd


# -----------------------------
# Helpers
# -----------------------------
def safe_slug(s: str, max_len: int = 80) -> str:
    raw = str(s)
    slug = re.sub(r"[^a-zA-Z0-9._-]+", "_", raw).strip("_")
    h = hashlib.md5(raw.encode("utf-8")).hexdigest()[:8]
    slug = slug[:max_len]
    return f"{slug}_{h}"

def get_context(code: str, line: int, radius: int = 3) -> Optional[str]:
    """Return a small snippet around 1-based line number."""
    if not isinstance(code, str) or not code:
        return None
    lines = code.splitlines()
    if not line or line < 1 or line > len(lines):
        return None
    start = max(1, line - radius)
    end = min(len(lines), line + radius)
    snippet = []
    for ln in range(start, end + 1):
        prefix = ">> " if ln == line else "   "
        snippet.append(f"{prefix}{ln:4d}: {lines[ln-1]}")
    return "\n".join(snippet)

def run_bandit(py_path: Path):
    proc = subprocess.run(
        ["bandit", "-f", "json", str(py_path)],
        capture_output=True,
        text=True
    )
    report = None
    try:
        if proc.stdout.strip():
            report = json.loads(proc.stdout)
    except Exception:
        report = None
    return report, proc.stderr, proc.returncode

def run_pylint(py_path: Path):
    proc = subprocess.run(
        ["pylint", str(py_path), "--output-format=json", "--score=no"],
        capture_output=True,
        text=True
    )
    issues = None
    try:
        if proc.stdout.strip():
            issues = json.loads(proc.stdout)
        else:
            issues = []
    except Exception:
        issues = None
    return issues, proc.stderr, proc.returncode


# -----------------------------
# Core: analyze ALL python files w/ issue context
# -----------------------------
def analyze_all_python_files_with_context(
    python_pr_df: pd.DataFrame,
    tmp_root: str = "tmp_python_issue_context_all",
    out_bandit_jsonl: str = "bandit_issue_level_with_context.jsonl",
    out_pylint_jsonl: str = "pylint_issue_level_with_context.jsonl",
    out_file_index_jsonl: str = "python_files_index_with_context.jsonl",
    limit_prs: int | None = None,
    context_radius: int = 3,
    max_issues_per_file: int | None = None,
):
    tmp_root = Path(tmp_root)
    tmp_root.mkdir(parents=True, exist_ok=True)

    bandit_out = Path(out_bandit_jsonl)
    pylint_out = Path(out_pylint_jsonl)
    file_out = Path(out_file_index_jsonl)

    df_iter = python_pr_df if limit_prs is None else python_pr_df.head(limit_prs)

    with bandit_out.open("w", encoding="utf-8") as fb, \
         pylint_out.open("w", encoding="utf-8") as fp, \
         file_out.open("w", encoding="utf-8") as ff:

        file_counter = 0
        bandit_issue_counter = 0
        pylint_issue_counter = 0

        for pr_idx, pr_row in df_iter.iterrows():
            html_url = pr_row.get("html_url")
            files = pr_row.get("files")

            if not isinstance(files, list):
                continue

            pr_dir = tmp_root / safe_slug(html_url or f"row_{pr_idx}")
            pr_dir.mkdir(parents=True, exist_ok=True)

            for f in files:
                if not isinstance(f, dict):
                    continue

                filename = f.get("filename")
                code = f.get("source_code")

                # only analyze .py with full source_code
                if not isinstance(filename, str) or not filename.endswith(".py"):
                    continue
                if not isinstance(code, str) or not code.strip():
                    continue

                # write file (single-file analysis keeps tooling simple)
                out_name = safe_slug(filename)
                py_path = pr_dir / f"{out_name}.py"
                py_path.write_text(code, encoding="utf-8")

                # run tools
                bandit_report, bandit_stderr, bandit_rc = run_bandit(py_path)
                pylint_issues, pylint_stderr, pylint_rc = run_pylint(py_path)

                # file index row
                ff.write(json.dumps({
                    "html_url": html_url,
                    "filename": filename,
                    "tmp_path": str(py_path),
                    "bandit_returncode": bandit_rc,
                    "pylint_returncode": pylint_rc,
                    "bandit_stderr": (bandit_stderr or ""),
                    "pylint_stderr": (pylint_stderr or ""),
                }, ensure_ascii=False) + "\n")
                file_counter += 1

                # ---- bandit: issue-level rows ----
                if isinstance(bandit_report, dict):
                    issues = bandit_report.get("results", []) or []
                    if max_issues_per_file is not None:
                        issues = issues[:max_issues_per_file]

                    for it in issues:
                        ln = it.get("line_number")
                        row = {
                            "html_url": html_url,
                            "filename": filename,
                            "tmp_path": str(py_path),

                            "tool": "bandit",
                            "test_id": it.get("test_id"),
                            "test_name": it.get("test_name"),
                            "severity": it.get("issue_severity"),
                            "confidence": it.get("issue_confidence"),
                            "line_number": ln,

                            "issue_text": it.get("issue_text"),
                            "more_info": it.get("more_info"),
                            "code_line": it.get("code"),
                            "context": get_context(code, ln, radius=context_radius),
                        }
                        fb.write(json.dumps(row, ensure_ascii=False) + "\n")
                        bandit_issue_counter += 1

                # ---- pylint: issue-level rows ----
                if isinstance(pylint_issues, list):
                    issues = pylint_issues
                    if max_issues_per_file is not None:
                        issues = issues[:max_issues_per_file]

                    for it in issues:
                        ln = it.get("line")
                        row = {
                            "html_url": html_url,
                            "filename": filename,
                            "tmp_path": str(py_path),

                            "tool": "pylint",
                            "type": it.get("type"),
                            "message_id": it.get("message-id"),
                            "symbol": it.get("symbol"),
                            "message": it.get("message"),

                            "line": ln,
                            "column": it.get("column"),
                            "endLine": it.get("endLine"),
                            "endColumn": it.get("endColumn"),

                            "context": get_context(code, ln, radius=context_radius),
                        }
                        fp.write(json.dumps(row, ensure_ascii=False) + "\n")
                        pylint_issue_counter += 1

        print("Done.")
        print("Files analyzed:", file_counter)
        print("Bandit issues:", bandit_issue_counter)
        print("Pylint issues:", pylint_issue_counter)
        print("Saved:",
              out_file_index_jsonl,
              out_bandit_jsonl,
              out_pylint_jsonl)


# run it
analyze_all_python_files_with_context(
    python_pr_df=python_pr_df,
    tmp_root="tmp_python_issue_context_all",
    out_bandit_jsonl="bandit_issue_level_with_context.jsonl",
    out_pylint_jsonl="pylint_issue_level_with_context.jsonl",
    out_file_index_jsonl="python_files_index_with_context.jsonl",
    limit_prs=None,
    context_radius=3,
    max_issues_per_file=None,
)
```
